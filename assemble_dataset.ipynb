```python
import pandas as pd
import numpy as np
import os
```

```python
main_dir = os.getcwd()
main_dir
```

```python
os.chdir('../../../datasets/sberbank-russian-housing-market')

df_train = pd.read_csv("train.csv")
df_macro = pd.read_csv("macro.csv")
df_train.head()
```

```python
#columns containing at least one na
na_col = list(df_train.loc[:,df_train.isna().sum(axis=0) != 0].columns)
for bad in na_col:
    print(bad)
```

```python
df_train.dropna().groupby("product_type")["life_sq"].mean()
```

```python
df_train.dropna().groupby("product_type")["life_sq"].median()
```

```python
# filling na with mean of that column based on the region and building type
df_train[na_col] = df_train.groupby(["sub_area", "product_type"])[na_col].transform(lambda x: x.fillna(x.dropna().mean()))
df_train[df_train["life_sq"].isna()]
```

```python
df_train[na_col] = df_train.groupby("product_type")[na_col].transform(lambda x: x.fillna(x.dropna().mean()))
df_train[df_train["life_sq"].isna()]
```

```python
sum(df_train.isna().sum(axis=0) != 0)
```

```python
final_columns = ["full_sq", "life_sq", "build_year", "product_type",
                 'max_floor', "num_room", "full_all", "work_all", 
                 "young_all", "children_school", '16_29_all', 'school_km',
                 'park_km', 'railroad_km', 'industrial_km', 
                 'metro_min_walk', 'metro_km_walk', 'oil_chemistry_km', 
                 'nuclear_reactor_km', 'radiation_km', 'university_km', 
                 'workplaces_km', 'shopping_centers_km', 'office_km', 
                 'big_church_km', 'church_synagogue_km', 'mosque_km', 
                 'theater_km','big_road1_km', 'big_road2_km', 
                 'market_shop_km','market_count_1000', 'sport_count_2000',
                 'church_count_3000', 'market_count_3000', 'timestamp',
                 'price_doc']
```

```python
df_train = df_train[final_columns]
df_train.shape
```

```python
df_macro.head()
```

```python
macro_na_col = list(df_macro.loc[:,df_macro.isna().sum(axis=0) != 0].columns)
macro_na_col
```

```python
df_macro.loc[:, macro_na_col]
```

```python
df_macro["month"] = df_macro['timestamp'].apply(lambda x: x.split("-")[1])
df_macro["year"] = df_macro['timestamp'].apply(lambda x: x.split("-")[0])
```

Cleaning up the data

```python
df_macro.loc[:, df_macro.dtypes == object].head()
```

```python
print(df_macro["child_on_acc_pre_school"].unique())
print(df_macro["modern_education_share"].unique())
print(df_macro["old_education_build_share"].unique())
```

```python
df_macro["child_on_acc_pre_school"] = df_macro["child_on_acc_pre_school"].replace("#!", np.nan)
df_macro["child_on_acc_pre_school"] = df_macro["child_on_acc_pre_school"].apply(lambda x: float(str(x).replace(",", "")))
```

```python
df_macro.drop(["modern_education_share", "old_education_build_share"], axis=1, inplace=True)
macro_na_col.remove("modern_education_share")
macro_na_col.remove("old_education_build_share")
```

```python
# replacing all na with mean of each column
df_macro[macro_na_col] = df_macro[macro_na_col].transform(lambda x: x.fillna(x.dropna().mean()))
```

```python
df_macro.head()
```

```python
df_macro.loc[:,df_macro.isna().sum(axis=0) != 0].head()
```

Replaced all the nas

```python
final_macro_col = ['timestamp','gdp_quart', 'cpi', 'ppi', 'unemployment',
                   'salary','income_per_cap', 'deposits_value', 
                   'mortgage_value']
df_macro = df_macro[final_macro_col]
```

```python
df = pd.merge(df_train, df_macro, on="timestamp")
df.head()
```

```python
os.chdir(main_dir)
df.to_csv("russian_housing.csv", index=False)
```
